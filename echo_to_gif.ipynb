# CT to gif

```python
import pydicom
import numpy as np
from PIL import Image
import os
import imageio.v2 as imageio
```

```python
files = [file for file in os.listdir('../echo') if 'raw' not in file]
```

```python
files.sort()
files
```

```python
for name in files:
    im = pydicom.dcmread(f'../echo/{name}')
    #ds = dcmread(...)
    rgb = pydicom.pixel_data_handlers.convert_color_space(im.pixel_array, "YBR_FULL", "RGB", per_frame=True)
    print(rgb.shape)
    #break
    #print(im.shape)
    #rows=int(im.Rows)
    #cols=int(im.Columns)
    #slices=len(im)
    #im = im.pixel_array.astype(float)
    #print(rows, cols, slices)
    #print(im)
    slices=rgb.shape[0]
    #print(im.shape[0], slices)
    
    #im = (np.maximum(im,0)/im.max())*255 # float pixels
    
    #im = np.uint8(im) # integers pixels
    #print(im.shape)
    if not os.path.exists(f'../echo_png/{name}'):
            os.mkdir(f'../echo_png/{name}')
    for slice in range(slices):
        #print(slice)
        final_image = Image.fromarray(rgb[slice, :, :, :])
        name = name.split('.')[0]
        
        final_image.save(f'../echo_png/{name}/slice_{slice}.png')

    
```

```python
for folder in os.listdir(f'../echo_png/'):
    print(folder)
    files = []
    images = []
    for file in os.listdir(f'../echo_png/{folder}'):
        #if 'raw' in file:
        filename = f'../echo_png/{folder}/{file}'
        files.append(filename)
    
    files.sort()
    #print(files)
    images = [imageio.imread(file) for file in files]
    #images.append(imageio.imread(filename))
        
    imageio.mimsave(f'../echo_gif/{folder}.gif', images, fps=3)
```
